```python
class Doctor:
    def __init__ (self, name: str, ID: int, phone: int, location: str, slots_available: list = generate_available_slots(), slots_reserved: list = []):
        self.name = name
        self.ID = ID
        self.phone = phone
        self.location = location
        self.slots_available = slots_available
        self.slots_reserved = slots_reserved

    def availability (self, day = 0):
        if day == 0:
            lines = [f"{s:%Y-%m-%d %H:%M} → {e:%H:%M}" for s, e in self.slots_available]
            return "\n".join(lines)
        else:
            lines = [f"{s:%H:%M} → {e:%H:%M}" for s, e in self.slots_available if s.date() == day]
            return "\n".join(lines)

D1 = Doctor("Mucci", 31540, 3450103547, "Milan")
D2 = Doctor("Grasso", 25471, 3460225471, "Milan")


class Patient:
    def __init__ (self, name: str, ID: int, phone: int, location: str, reservations: list = []):
        self.name: str = name
        self.ID: int = ID
        self.phone: int = phone
        self.location: str = location
        self.reservations: list = reservations

P1 = Patient ("Ciacio", 13209, 346812072, "Milan")


class BookingService:
    def __init__ (self):
        pass

    def book (self, patient, doctor, slot):
        if slot in doctor.slots_reserved:
            raise ValueError("Slot already reserved")
        elif slot not in doctor.slots_reserved and slot not in doctor.slots_available:
            raise ValueError("Slot not available")
        else:
            doctor.slots_available.remove(slot)
            doctor.slots_reserved.add(slot)
            patient.reservations.add(slot)
            return "Booking completed"
        
       
        
print (D1.slots_available)
print(D1.slots_reserved)
print(P1.reservations)

BookingService.book(P1, D1, date(2025,11,2,))











#print(P1.reservations)
#print(D1.availability(date(2025, 10, 31)))   # example of imput: date(2025,11,2)




##############################################################################################################
# Functions

from datetime import datetime, date, time, timedelta
def generate_available_slots (opening_time = time(8,0), closing_time = time(15,00), slot_length = timedelta(minutes=30)):
    slots = []
    t = datetime.now()
    while t <= datetime.now() + timedelta(days = 7):
        if t.time() < opening_time:
            t = datetime.combine (t.date(), opening_time)
        if t.time() > closing_time:
            t = datetime.combine ((t + timedelta(days=1)).date(), opening_time) 
        slots.append((t, t + slot_length))
        t += slot_length
    return slots
```
